```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv("Downloads/Toyota.csv", index_col= 0, na_values =["??" , "????"] )
print(cars_data)
```

```python
cars_data.dropna(axis=0,inplace=True)
print(cars_data)
```

```python
plt.scatter(cars_data['Age'],cars_data['Price'], c ='red')
plt.title('scatter plot of Price vs Age of the cars')
plt.xlabel('Age(month)')
plt.ylabel('Price(euros)')
plt.show()
```

```python
sns.set(style ="darkgrid")
sns.regplot(x=cars_data['Age'], y=cars_data['Price'],fit_reg=False)
```

```python
sns.distplot(cars_data['Price'])
```

```python
sns.lmplot(x='Age',y='Price',data=cars_data, hue= 'FuelType', legend=True, palette="Set1")
```

```python
plt.scatter(cars_data['KM'],cars_data['Price'], c ='blue')
plt.title('scatter plot of Price vs KM of the cars')
plt.xlabel('KM(kilometers)')
plt.ylabel('Price(euros)')
plt.show()
```

```python
sns.boxplot(y=cars_data["Price"])
```

```python
sns.set(style ="darkgrid")
sns.regplot(x=cars_data['KM'], y=cars_data['Price'],fit_reg = False)
```

```python
sns.distplot(cars_data['KM'])
```

```python
p=sns.lmplot(x='Age',y='Price',data=cars_data, hue= 'MetColor', legend=True, palette="Set1")
print(p)
```

```python
sns.displot(cars_data['Age'])
```

```python
sns.countplot(x="FuelType",data = cars_data)
```
